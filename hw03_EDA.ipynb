### Задание 13  (0.5 балла)

Представьте все порядковые и бинарные признаки в виде числовых значений {0, ..., n - 1}, где n это количество значений признака. Избавьтесь от категориальных признаков.

```python
cat_columns = ['credit_history', 'purpose', 'personal_status_sex', 'other_debtors',
               'property', 'other_installment_plans', 'housing']

data.drop(columns=cat_columns, inplace=True)
```

```python
status_mapping = {'no checking account': 0,
                  '... < 0 DM': 1,
                  '... >= 200 DM / salary for at least 1 year': 2,
                  '0<= ... < 200 DM': 3}

savings_mapping = {'unknown/no savings account': 0,
                   '... <  100 DM': 1,
                   '100 <= ... <  500 DM': 2,
                   '... >= 1000 DM': 3,
                   '500 <= ... < 1000 DM': 4}

employment_duration_mapping = {'unemployed': 0,
                               '< 1 yr': 1,
                               '1 <= ... < 4 yrs': 2,
                               '4 <= ... < 7 yrs': 3,
                               '>= 7 yrs': 4}

installment_rate_mapping = {'< 20': 0,
                            '25 <= ... < 35': 1,
                            '20 <= ... < 25': 2,
                            '>= 35': 3}

present_residence_mapping = {'< 1 yr': 0,
                             '1 <= ... < 4 yrs': 1,
                             '4 <= ... < 7 yrs': 2,
                             '>= 7 yrs': 3}

number_credits_mapping = {'1': 0,
                          '2-3': 1,
                          '4-5': 2,
                          '>= 6': 3}

job_mapping = {'unemployed/unskilled - non-resident': 0,
               'unskilled - resident': 1,
               'skilled employee/official': 2,
               'manager/self-empl./highly qualif. employee': 3}

people_liable_mapping = {'0 to 2': 0, '3 or more': 1}

telephone_mapping = {'no': 0, 'yes (under customer name)': 1}

foreign_worker_mapping = {'no': 0, 'yes': 1}

credit_risk_mapping = {'bad': 0, 'good': 1}

data.replace({'status': status_mapping,
              'savings': savings_mapping,
              'employment_duration': employment_duration_mapping,
              'installment_rate': installment_rate_mapping,
              'present_residence': present_residence_mapping,
              'number_credits': number_credits_mapping,
              'job': job_mapping,
              'people_liable': people_liable_mapping,
              'telephone': telephone_mapping,
              'foreign_worker': foreign_worker_mapping,
              'credit_risk': credit_risk_mapping}, inplace=True)

data
```

```python
def predict_helper(features: np.array) -> int:
    # age < 0.008 * (amount - 7000)
    if features[1] < 0.008 * (features[2] - 7000):
        return 0
    # age < 1.3 * (duration - 15) and status == 'no checking account'
    if features[7] < 1.3 * (features[1] - 15) and features[0] == 0:
        return 0
    # age < 0.007 * (amount - 5_750)
    if features[7] < 0.007 * (features[2] - 5_750):
        return 0
    return 1
```

```python
def predict(Xtest: np.array) -> np.array:
    return np.apply_along_axis(func1d=predict_helper, axis=1, arr=Xtest)
```

```python
accuracy_score(y_train, predict(X_train))
```

На тренировочных данных новый классификатор показывает результат лучше, чем константный, значит, некоторые закономерности, которые были в тренировочных данных нам удалось зафиксировать. Проверим, насколько хорошо "модель" будет работать на новых данных.
